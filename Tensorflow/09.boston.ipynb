```python
import pandas as pd
```

```python
df = pd.read_csv('data/housing.csv', delim_whitespace=True,\
                 header=None)

print(df.info())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import tensorflow as tf
```

```python
# Seed 값 생성
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
# 데이터 입력
dataset = df.values
X = dataset[:,0:13].astype(float)
y = dataset[:,13]
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1)) # 선형회귀이기 때문에 마지막에 activation이 없음 나오는 값 그대로가 가격

# compile
model.compile(loss='mean_squared_error', \
             optimizer='adam')
model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
#evaluate: 정확도 loss 값 평가 VS predict: 입력값과 출력값을 뱉어주기만 하는 역할
y_prediction = model.predict(X_test).flatten()
```

```python
print(type(y_prediction))
```

```python
print(y_test)
```

```python
for i in range(len(y_test)):
    label = y_test[i]
    prediction = y_prediction[i]
    print("실제가격: {:.3f}, 예상가격: {:.3f}".\
         format(label, prediction))
    
```
